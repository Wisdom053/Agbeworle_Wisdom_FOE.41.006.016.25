```python

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self, ):
        print(f"name:", self.name)
        print(f"age:", self.age)


#child class: student
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self, course_name):
        print(f"Student {self.name} has successfully enrolled in {course_name}.")
        self.course_name = course_name

    def display_info(self):
        super().display_info()
        print(f"Student_id:{self.student_id}")


class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course_name):
        print(f"Lecturer{self.name} is now teaching{course_name}")
        self.course_name = course_name


if __name__ == "__main__":
    print("----creating objects----")

    student1 = Student(name="Wisdom", age=35, student_id="FOE.41.006.016.25")
    lecturer1 = Lecturer(name="DR.Cobbinah", age=95, employee_id="UMaT.22222")

    print("\n--- Demonstrating student class---")
    student1.display_info()
    student1.enroll_course("object oriented programming in python")

    print("\n--- Demonstrating lecturer class---")
    lecturer1.display_info()
    lecturer1.teach_course("Object oriented programming in python")


```
